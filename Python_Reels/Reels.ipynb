Cleaning of content_type  & primary_platform 


```python
platforms = [
    'TikTok', 'Instagram Reels', 'YouTube Shorts',
    'Facebook Reels', 'Snapchat Spotlight'
]

def clean_platforms(platform):
    items = str(platform).split(' | ')
    items = [i for i in items if i in platforms]
    return ' | '.join(items)

Reels['primary_platform'] = Reels['primary_platform'].apply(clean_platforms)


```

```python
Reels.dtypes
```

```python
Reels['primary_platform'] = Reels['primary_platform'].replace('', pd.NA)
```

```python

Reels['primary_platform'] = Reels['primary_platform'].fillna('Other')
```

```python
df_platform = (
    Reels.assign(primary_platform=Reels["primary_platform"].str.split("|"))
         .explode("primary_platform")
)

df_platform["primary_platform"] = df_platform["primary_platform"].str.strip()

df_platform = df_platform[["user_id", "primary_platform"]].reset_index(drop=True)
```
